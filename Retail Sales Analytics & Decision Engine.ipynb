```python
import numpy as np
import csv

class SalesAnalyzer:
    
    def __init__(self, filename):
        
        self.products = []
        self.units = []
        self.prices = []
        self.locations = []
        self.genders = []
        
        try:
            with open(filename,'r') as file:
                reader = csv.DictReader(file)
                
                for row in reader:
                    self.products.append(row["Product"])
                    self.units.append(int(row["Units_Sold"]))
                    self.prices.append(float(row["Price"]))
                    self.locations.append(row["Location"])
                    self.genders.append(row["Gender"])
                    
        except Exception as e:
            print("Error loading file:", e)
            return
        
        # Convert to NumPy arrays
        self.products = np.array(self.products)
        self.units = np.array(self.units)
        self.prices = np.array(self.prices)
        self.locations = np.array(self.locations)
        self.genders = np.array(self.genders)
        self.revenue = self.units * self.prices

   
    # KPI SECTION
    def total_revenue(self):
        total = np.sum(self.revenue)
        print(f"\n Total Revenue : ₹{total:,.2f}")
        return total

    def average_order_value(self):
        aov = np.mean(self.revenue)
        print(f" Average Order Value (AOV): ₹{aov:,.2f}")
        return aov

    # PRODUCT INSIGHTS
    def sales_by_product(self):
        print("\n Sales by Product")
        products = np.unique(self.products)
        
        for p in products:
            mask = self.products == p
            total = np.sum(self.units[mask])
            print(f"{p:<12} : {total} units")

    def top_product(self):
        products, counts = np.unique(self.products, return_counts=True)
        top = products[np.argmax(counts)]
        print(f"\n Top Selling Product: {top}")
        return top

    def low_performing_products(self):
        products, counts = np.unique(self.products, return_counts=True)
        low = products[counts < 2]
        
        print("\n Low Performing Products:")
        for p in low:
            print(p)
        
        return low

    
    # SEGMENT INSIGHTS
    def sales_by_location(self):
        print("\n Sales by Location")
        locations = np.unique(self.locations)

        for loc in locations:   
            mask = self.locations == loc
            total = np.sum(self.units[mask])
            print(f"{loc:<12} : {total} units")

    def sales_by_gender(self):
        print("\n Sales by Gender")
        genders = np.unique(self.genders)
        
        for g in genders:
            mask = self.genders == g
            total = np.sum(self.units[mask])
            print(f"{g:<10} : {total} units")

    #  BUSINESS DECISIONS
    def discount_strategy(self):
        revenue = np.sum(self.revenue)
        
        print("\n Discount Strategy:")
        
        if revenue < 100000:
            print("Apply 20% discount (Low performance)")
        elif revenue < 300000:
            print("Apply 10% discount (Moderate performance)")
        else:
            print("No discount needed (Good performance)")

    def restock_recommendation(self):
        top = self.top_product()
        print(f"\n Restock Recommendation: Increase stock for '{top}'")

    # FULL BUSINESS REPORT
    # =========================

    def full_report(self):
        print("\n" + "="*50)
        print(" FULL BUSINESS REPORT")
        print("="*50)

        self.total_revenue()
        self.average_order_value()
        self.sales_by_product()
        self.top_product()
        self.low_performing_products()
        self.sales_by_location()
        self.sales_by_gender()
        self.discount_strategy()
        self.restock_recommendation()
```
